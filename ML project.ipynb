```python
#logistic algorthim
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("pregnancy_Nutrition.csv")
```

```python
dataset
```

```python
le = LabelEncoder()
dataset["Risk"] = le.fit_transform(dataset["Risk"])
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)

```

```python
independent = dataset[["Age","BMI","Hemoglobin","BloodPressure","SugarLevel","ProteinIntake"]]
```

```python
independent
```

```python
dependent = dataset["Risk"]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
# creating logistic Algorithm Classification Model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
from sklearn.metrics import classification_report
class_report = classification_report(y_test,y_pred)
print(class_report)
```
